```python
print(f'–ú—É–∂—á–∏–Ω –Ω–∞ –º–µ–Ω–µ–¥–∂–º–µ–Ω—Ç–µ –±–æ–ª—å—à–µ —á–µ–º –Ω–∞ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–µ –≤ {manage_male_cnt / marketing_male_cnt} —Ä–∞–∑')
```
